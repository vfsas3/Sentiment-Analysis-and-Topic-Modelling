```python
import numpy as np
import pandas as pd
import seaborn as sns
import re
```

```python
df= pd.read_csv("Vaccine_dataset_AstraZeneca.csv")
df
```

```python
df.isna().sum()
```

```python
df = df.fillna(0)
df
```

```python
def extract_tokens(sentence):
    return re.findall(r'\b\w+\b', sentence)

# Criar uma nova coluna com os tokens (palavras)
df['tokens'] = df['Tweet'].apply(extract_tokens)

df['tokens'] = df['tokens'].apply(lambda tokens: [
    re.sub('[,|.|?]', '', re.sub('-', ' ', re.sub("n\'t", " not", word.lower())))
    for word in tokens
])

df
```

StemizaÃ§Ã£o

```python
from nltk import stem
# Define Porter Stemmer
porter = stem.porter.PorterStemmer()
# Define Snowball Stemmer
snowball = stem.snowball.EnglishStemmer()
# Define Lancaster Stemmer
lancaster = stem.lancaster.LancasterStemmer()
```

```python
df['porter_stemmed'] = df['tokens'].apply(lambda tokens: [porter.stem(word) for word in tokens])
df['snowball_stemmed'] = df['tokens'].apply(lambda tokens: [snowball.stem(word) for word in tokens])
df['lancaster_stemmed'] = df['tokens'].apply(lambda tokens: [lancaster.stem(word) for word in tokens])
print('Porter Stemmer:', df['porter_stemmed'])
print('Snowball Stemmer:', df['snowball_stemmed'])
print('Lancaster Stemmer:', df['lancaster_stemmed'])
```
